# Load data frames for visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Reduce the data size by taking a random sample to visualize

```python
pop=df4
pop = pop.sample(n=25, random_state=121)
pop.columns
```

```python
pop=pop.sort_values(by='averagerating',ascending=True)
```

# calculating the sum of both gross and domestic gross while checking if they both relate to the sell of the movie

```python
genre_grouped = pop.groupby('genre')[['domestic_gross', 'foreign_gross']].sum()
```

```python
genre_grouped.plot(kind='bar', figsize=(12, 6))
plt.title('Domestic and Foreign Gross by Genre')
plt.xlabel('Genre')
plt.ylabel('Gross Earnings')
plt.show()
```

We further want to understand if Domestic gross does affect the revenue with a scatterplot.

```python
sns.scatterplot(x='averagerating',y='domestic_gross',data=pop)
plt.title('Ratings on gross')
```

The above graph Shows a weak correlation

# Finding the most popular movie, movies by title and genre


```python
pop[pop['domestic_gross'].max()==pop ['domestic_gross']]
```

```python
Top10 = pop.nlargest(10, 'averagerating')[['title','genre', 'averagerating', 'studio']]\
.set_index('title','genre')
Top10
```

```python
custom_palette = sns.color_palette("Set1", len(Top10['genre'].unique()))
sns.barplot(x='averagerating', y=Top10.index,data=Top10,hue='genre',palette=custom_palette)
plt.legend(title='Genre', title_fontsize='20', loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Top ten movies_one')
```

Comparison between top ten movies on graph one and two( Group1_Top10 that was before dropping null values).
This was to be used to compare and further recommend what genres to produce.

```python
custom_palette = sns.color_palette("Set1", len(Top10['genre'].unique()))
sns.barplot(x='averagerating', y=Group1_Top10.index,data=Group1_Top10,hue='genres',palette=custom_palette)
plt.legend(title='Genres', title_fontsize='20', loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Top ten movies-two')
```

```python
Group1_Top10.rename (columns = {"genres" : "genre"}, inplace = True) 
```

```python
genre_counts_Top10 = Top10['genre'].str.split(',', expand=True).stack().value_counts()
genre_counts_Group1_Top10 = Group1_Top10['genre'].str.split(',', expand=True).stack().value_counts()
```

```python
genre_counts = pd.concat([genre_counts_Top10, genre_counts_Group1_Top10], axis=1).fillna(0)
genre_counts.columns = ['Dataset 1', 'Dataset 2']
```

```python
genre_counts.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.title('Comparison of top ten Geners ')
plt.legend(title='Dataset')

plt.show()
```

By comparing the two data set i have managed analyze the success of movies according to their producers and genre to recommend.That is:
